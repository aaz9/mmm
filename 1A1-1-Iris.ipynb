```python
q1 = np.percentile(data['sepal.width'], 25)
q3 = np.percentile(data['sepal.width'], 75)
IQR = q3 - q1

ll = q1 - (IQR*1.5)
ul = q3 + (IQR*1.5)

lo = data[data['sepal.width'] < ll].index.tolist()
uo = data[data['sepal.width'] > ul].index.tolist()
bi = list(set(lo+uo))

bb = True

if bb:
    data.drop(bi, inplace = True, errors = "ignore")
    print("outliers from {} feature remove".format(data['sepal.width']))
```
